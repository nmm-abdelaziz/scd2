```python
from urllib.request import urlretrieve

url = ("https://data-api.ecb.europa.eu/service/data/BPS?format=csvdata")
filename = "/lakehouse/default/Files/bronze/data.csv"
path, headers = urlretrieve(url, filename)
```

```python
facts_df = (
    spark.read.option("header", True)
    .option("inferSchema", True)
    .option("sep",",")
    .csv("Files/bronze/data.csv")
)
```

```python
fact_annual = facts_df.where(facts_df.FREQ=="A")
fact_quarterly = facts_df.where(facts_df.FREQ=="Q")
fact_monthly = facts_df.where(facts_df.FREQ=="M")
```

```python
fact_annual.write.mode("overwrite").format("delta").save("Files/silver/fact_annual_inter_invest")
fact_quarterly.write.mode("overwrite").format("delta").save("Files/silver/fact_quarterly_inter_invest")
fact_monthly.write.mode("overwrite").format("delta").save("Files/silver/fact_monthly_inter_invest")
```
